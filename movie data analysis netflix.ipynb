# Exploration Summary üöÄ
Our dataframe contains 9827 rows and 9 columns ‚Äî a solid sample size for strong analysis.

The dataset is tidy: no missing values, no duplicates. Pre-processing will be smooth.

Release_Date column needs conversion to datetime format for accurate chronological analysis.

Columns like Overview, Original_Language, and Poster-Url are currently less useful for quantitative analysis.

There are noticeable outliers in the Popularity column ‚Äî deeper exploration required.

Vote_Average can be binned or categorized to uncover hidden patterns and audience clusters.

The Genre column has comma-separated values and extra whitespaces; cleaning is needed for proper grouping and analysis.



```python
# Data Cleaning Steps

# 1. Remove duplicate rows
df = df.drop_duplicates()

# 2. Handle missing values (if any, for reference)
df = df.dropna()  # If you want to drop all rows with any NaN values
# Or, fill with suitable value:
# df = df.fillna({'col_name': value})

# 3. Convert Release_Date to datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# 4. Remove irrelevant columns for analysis (if not needed)
df = df.drop(['Overview', 'Original_Language', 'Poster_Url\r'], axis=1)

# 5. Remove outliers in Popularity (optional: using IQR method)
Q1 = df['Popularity'].quantile(0.25)
Q3 = df['Popularity'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Popularity'] >= Q1 - 1.5*IQR) & (df['Popularity'] <= Q3 + 1.5*IQR)]

# 6. Bin/categorize Vote_Average
df['Vote_Group'] = pd.cut(df['Vote_Average'], bins=[0,5,7,10], labels=['Low','Medium','High'])

# 7. Clean Genre column: remove extra spaces and split into list
df['Genre'] = df['Genre'].str.replace(' ', '').str.split(',')

# Optional: check changes
df.head()
```

1. What is the most frequent genre of movies released on Netflix?

```python
# Flatten genre list and count frequency
from collections import Counter
genre_flat = [g for sublist in df['Genre'] for g in sublist]
most_common_genre = Counter(genre_flat).most_common(1)[0]
print("Most frequent genre:", most_common_genre[0], "| Frequency:", most_common_genre[1])
```

2. What genres have highest votes?

```python
# Sum vote counts per genre
genre_votes = {}

for genres, votes in zip(df['Genre'], df['Vote_Count']):
    for g in genres:
        genre_votes[g] = genre_votes.get(g, 0) + votes

# Top genres by votes
top_genres_votes = sorted(genre_votes.items(), key=lambda x: x[1], reverse=True)[:5]
print("Genres with highest votes:", top_genres_votes)
```

3. What movie got the highest popularity? What's its genre?

```python
# Get row with highest popularity
max_pop_row = df.loc[df['Popularity'].idxmax()]
print("Highest popularity movie:", max_pop_row['Title'])
print("Its genres:", max_pop_row['Genre'])
print("Popularity score:", max_pop_row['Popularity'])
```

4. What movie got the lowest popularity? What's its genre?

```python
min_pop_row = df.loc[df['Popularity'].idxmin()]
print("Lowest popularity movie:", min_pop_row['Title'])
print("Its genres:", min_pop_row['Genre'])
print("Popularity score:", min_pop_row['Popularity'])
```

5. Which year has the most filmed movies?

```python
# Release_Year column should be present from previous extraction
most_filmed_year = df['Release_Year'].value_counts().idxmax()
num_movies = df['Release_Year'].value_counts().max()
print("Year with most filmed movies:", most_filmed_year, "| Number of movies:", num_movies)
```

6. Is there a correlation between popularity and average rating (Vote_Average)?

```python
corr = df['Popularity'].corr(df['Vote_Average'])
print(f"Correlation between Popularity and Vote_Average: {corr:.2f}")
```

Which genre is most popular in specific years? (example for top genre per year)

```python
# Group by year, explode genres, count
df_exploded = df.explode('Genre')
top_genres_by_year = df_exploded.groupby(['Release_Year', 'Genre']).size().reset_index(name='Count')
most_popular_genre_by_year = top_genres_by_year.loc[top_genres_by_year.groupby('Release_Year')['Count'].idxmax()]
print(most_popular_genre_by_year[['Release_Year', 'Genre', 'Count']])
```

 Which movies had low ratings but high popularity? (Outlier detection)

```python
# Define "low rating" and "high popularity" thresholds
low_rating = df['Vote_Average'].quantile(0.25)
high_popularity = df['Popularity'].quantile(0.75)
outlier_movies = df[(df['Vote_Average'] <= low_rating) & (df['Popularity'] >= high_popularity)]
print("Movies with low ratings but high popularity:\n", outlier_movies[['Title', 'Genre', 'Popularity', 'Vote_Average']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df['Popularity'].corr(df['Vote_Average'])
print(f"Correlation between Popularity and Vote_Average: {corr:.2f}")

plt.figure(figsize=(7,5))
sns.heatmap(df[['Popularity', 'Vote_Average', 'Vote_Count']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

```

```python
from sklearn.cluster import KMeans
import numpy as np

# Only use numeric columns for clustering
X = df[['Popularity', 'Vote_Count', 'Vote_Average']].dropna()

kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X)

df['Cluster'] = clusters

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Popularity'], y=df['Vote_Count'], hue=df['Cluster'], palette='Set2', alpha=0.7)
plt.title("Movie Clusters by Popularity & Vote Count")
plt.xlabel("Popularity")
plt.ylabel("Vote Count")
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Create multi-plot dashboard style
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Pie Chart: Genre distribution
genre_flat = [g for genre_list in df['Genre'] for g in genre_list]
top_genre_counts = Counter(genre_flat).most_common(7)
genres, counts = zip(*top_genre_counts)
colors = sns.color_palette("pastel")
axes[0,0].pie(counts, labels=genres, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor':'white'})
axes[0,0].set_title("Genre Distribution (Top 7)")

# 2. Line Plot: Movies released per year
year_counts = df['Release_Year'].value_counts().sort_index()
sns.lineplot(y=year_counts.values, x=year_counts.index, marker='o', ax=axes[0,1], color='teal')
axes[0,1].set_title("Movies Released per Year")
axes[0,1].set_xlabel("Year")
axes[0,1].set_ylabel("Number of Movies")

# 3. Scatter Plot: Popularity vs Average Rating (clusters if available)
if 'Cluster' in df.columns:
    sns.scatterplot(x=df['Popularity'], y=df['Vote_Count'], hue=df['Cluster'], palette='Set2', alpha=0.7, ax=axes[1,0])
    axes[1,0].set_title("Clusters: Popularity & Votes")
else:
    sns.scatterplot(x=df['Vote_Average'], y=df['Popularity'], alpha=0.7, ax=axes[1,0])
    axes[1,0].set_title("Popularity vs Average Rating")

axes[1,0].set_xlabel("Popularity")
axes[1,0].set_ylabel("Vote Count")

# 4. Bar Plot: Average Popularity per Genre
df_exploded = df.explode('Genre')
avg_popularity = df_exploded.groupby('Genre')['Popularity'].mean().sort_values(ascending=False).head(7)
sns.barplot(x=avg_popularity.index, y=avg_popularity.values, palette='mako', ax=axes[1,1])
axes[1,1].set_title("Avg Popularity by Genre")
axes[1,1].set_ylabel("Avg Popularity")
axes[1,1].set_xticklabels(avg_popularity.index, rotation=45)

plt.tight_layout()
plt.show()
```
